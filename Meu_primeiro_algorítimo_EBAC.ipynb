## Tabela Verdade


```python
print(10 > 2 and 10 > 2) # V and V - V
print(10 > 2 and 10 < 2) # V and F - F
print(10 < 2 and 10 > 2) # F and V - F
print(10 < 2 and 10 < 2) # F and F - F
```

```python
print(10 > 2 or 10 > 2) # V or V - V
print(10 > 2 or 10 < 2) # V or F - V
print(10 < 2 or 10 > 2) # F or V - V
print(10 < 2 or 10 < 2) # F or F - F
```

```python
not 10 > 2 # V - F
```

```python
not 10 < 2 # F - V
```

# Estrutura de Repetição
## for

```python
# for range
for contador in range(10):
  print(contador)
```

```python
# for range + break
for contador in range(100):
  print(contador)
  if contador == 7:
    break
```

```python
lista = ["num1", "num2"]
```

```python
for val in lista:
  print(val)
```

```python
# for com enumerate
for i, val in enumerate(lista):
  print(i, val)
```

# for com operadores compostos

```python
lista2 = [2, 3, 4]
```

```python
for val in lista2:
  val += 1
  print(val)
```

```python
for val in lista2:
  val *= 2
  print(val)
```

# while

```python
val = 1
while (val < 6):
  val += 2
  print(val)
```

```python
val = 10
while val < 6:
  val += 2
else:
  print("O valor de val é: ",val)
```
